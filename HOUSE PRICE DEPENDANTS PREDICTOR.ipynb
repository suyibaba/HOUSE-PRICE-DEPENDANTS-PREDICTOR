### TEST TRAIN DATA

```python
from sklearn.model_selection import train_test_split
```

```python
#random is done to make sure we get the same value every time
x_train, x_test, y_train, y_test = train_test_split(x_multi,y_multi, test_size=0.2, random_state = 0)
```

```python
x_test.shape, x_train.shape , y_train.shape , y_test.shape
```

```python
#create an object
lm_a = LinearRegression()
```

```python
lm_a.fit(x_train,y_train)
```

```python
#to predict the test dependent variable, so i will give my best independent variable
y_test_a = lm_a.predict(x_test)
y_test_a
```

```python
y_train_a = lm_a.predict(x_train)
y_train_a
```

```python
#to get the R squared value for my training and test data
from sklearn.metrics import r2_score
```

```python
#y_test_a is the predicted value
r2_score(y_test, y_test_a)
```

```python
#r2 value for our training set
r2_score(y_train, y_train_a)
```

### MULTIPLE LINEAR REGRESSION PREDICTION USING RIDGE REGRESSION AND LASSO

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler().fit(x_train)
```

```python
#name a scaler varaible
x_train_s = scaler.transform(x_train)
```

```python
x_test_s = scaler.transform(x_test)
```

```python
#first we import ridge from sklearn
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.5)
lm_r.fit(x_train_s,y_train)
```

```python
r2_score(y_test, lm_r.predict(x_test_s))
```
